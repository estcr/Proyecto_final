```python
import os
from dotenv import load_dotenv
import pinecone
import pandas as pd
from openai import OpenAI
from typing import List

# Load environment variables
load_dotenv()
api_key_pinecone = os.getenv("apipinecone")
api_key_openai = os.getenv("apigpt_key")

# Initialize OpenAI client
client = OpenAI(api_key=api_key_openai)

# Initialize Pinecone
pc = pinecone.Pinecone(api_key=api_key_pinecone)

# Create index if it doesn't exist
if 'tuguia' not in pc.list_indexes().names():
    pc.create_index(
        name='tuguia',
        dimension=1536,  # OpenAI text-embedding-3-small has 1536 dimensions
        metric='cosine',
        spec=pinecone.ServerlessSpec(
            cloud='aws',
            region='us-east-1'
        )
    )

index = pc.Index('tuguia')

def generate_embeddings(texts: List[str]) -> List[List[float]]:
    """
    Generate embeddings using OpenAI's API
    
    Args:
        texts (List[str]): List of texts to generate embeddings for
        
    Returns:
        List[List[float]]: List of embedding vectors
    """
    embeddings = []
    
    # Process texts in batches to avoid rate limits
    batch_size = 100  # Adjust based on your needs
    
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i + batch_size]
        response = client.embeddings.create(
            model="text-embedding-3-small",
            input=batch,
            encoding_format="float"
        )
        # Extract embeddings from response
        batch_embeddings = [data.embedding for data in response.data]
        embeddings.extend(batch_embeddings)
    
    return embeddings

# Combine fields for embedding
texts = df["Actividad"] + " " + df["Ubicación"] + " " + df["Descripción"]

# Generate embeddings
embeddings = generate_embeddings(texts)

# Create vectors for Pinecone
vectors = [
    {
        "id": f"vec{i+1}",
        "values": embedding,
        "metadata": {
            "Actividad": df.loc[i, "Actividad"],
            "Ubicación": df.loc[i, "Ubicación"],
            "Descripción": df.loc[i, "Descripción"]
        }
    }
    for i, embedding in enumerate(embeddings)
]

# Upsert vectors to Pinecone
index.upsert(
    vectors=vectors,
    namespace="ns1"
)

print("Vectors successfully inserted into Pinecone.")
```
